```python
from google import genai
import os
from IPython.display import display, Markdown

# Configure API key (replace with your actual key or use environment variable)
GOOGLE_API_KEY = os.getenv("GEMINI_API_KEY", "YOUR_API_KEY_HERE")

# Initialize the client
client = genai.Client(api_key=GOOGLE_API_KEY)

# Set the model to use
MODEL_ID = "gemini-2.5-flash" 

# Define the prompt with system instructions and user query
prompt = """Objective: Generate multi-format alerts for suspicious transactions. Generate both SMS and email alerts for a customer whose card was used in a foreign location. Ensure urgency and clarity.
Context: A customer's card was used in a foreign location, which is unusual for their spending patterns. The transaction occurred at 3 AM local time, and the amount was $500. The customer has not traveled recently.
Instructions:
1. Create an SMS alert that is concise and emphasizes the urgency of the situation.
2. Create an email alert that provides more details about the transaction and instructions on what the customer should do next.
3. Ensure that both alerts are clear and easy to understand, prompting the customer to take immediate action if the transaction was unauthorized.
4. Use a tone that conveys concern and encourages the customer to verify the transaction.

Generate the SMS and email alerts based on the above information."""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt
)

display(Markdown("### Prompt Response:"))
display(Markdown(response.text))
```
