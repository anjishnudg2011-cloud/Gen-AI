```python
import requests
import os
from dotenv import load_dotenv
load_dotenv(override=True)  # override=True to reload variables if already loaded
# Get API key from environment variable
gemini_api_key = os.getenv("GEMINI_API_KEY")

if gemini_api_key is None:
    raise ValueError("GEMINI_API_KEY not found in .env file")

# Set up the API endpoint
url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-3-flash-preview:generateContent"

# Configure headers
headers = {
    "Content-Type": "application/json",
    "x-goog-api-key": gemini_api_key
}
# Create the prompt payload
prompt = """Objective: We are trying to automate transaction analysis using AI.
Role: You are a transaction analyst.
Task: Read the following CSV file of transactions and categorize them into Food, Travel, Bills, Shopping.

Date,Description,Amount
2026-01-03,Swiggy Food Order,450
2026-01-04,Uber Ride to Office,210
2026-01-05,Amazon Purchase - Headphones,1299
2026-01-06,BESCOM Electricity Bill,850
2026-01-07,Zomato Dinner,520
2026-01-08,Metro Smart Card Recharge,300
2026-01-09,Flipkart - T-Shirt,499
2026-01-10,Airtel Mobile Bill,499
2026-01-11,Big Bazaar Grocery,780
2026-01-12,OLA Airport Ride,650
2026-01-13,Dominos Pizza,630
2026-01-14,IRCTC Train Ticket,1240
2026-01-15,Spencers Grocery Store,340
2026-01-16,Myntra - Shoes,1899
2026-01-17,HP Gas Cylinder Booking,985
2026-01-18,Swiggy Breakfast,270
2026-01-19,Petrol Bunk Fuel,1200
2026-01-20,Uber Outstation Trip,2200
2026-01-21,Tata Play DTH Recharge,350
2026-01-22,Amazon - Mobile Cover,199"""
payload = {
    "contents": [
        {
            "parts": [
                {
                    "text": prompt
                }
            ]
        }
    ],
    "generationConfig": {
        "temperature": 0.5,
        "maxOutputTokens": 1500
    }
}
# Send the request
response = requests.post(url, json=payload, headers=headers)

response_data = response.json()
#print("\nUser prompt:\n",prompt)
#print("\n" + "="*60 + "\n")
# Extract the text from the response
if 'candidates' in response_data:
    content = response_data['candidates'][0]['content']['parts'][0]['text']
    print("AI Response:")
    print(content.strip())
```
