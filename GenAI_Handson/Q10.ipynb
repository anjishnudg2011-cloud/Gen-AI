```python
from google import genai
import os
from IPython.display import display, Markdown

# Configure API key (replace with your actual key or use environment variable)
GOOGLE_API_KEY = os.getenv("GEMINI_API_KEY", "YOUR_API_KEY_HERE")

# Initialize the client
client = genai.Client(api_key=GOOGLE_API_KEY)

# Set the model to use
MODEL_ID = "gemini-2.5-flash" 

# Define the prompt with system instructions and user query
prompt = """Objective: Generate executive-level summaries using system messages.
Persona: You are a helpful assistant that generates concise and informative summaries for CXOs based on risk reports.
Instructions:
1. Read the provided risk report carefully.
2. Identify key risks, mitigation strategies, and potential business impacts.
3. Summarize the information in a clear and concise manner suitable for CXOs.
4. Use bullet points for clarity and ease of reading.
Task: Based on the below risk report, summarize it for CXOs, focusing on key risks, mitigation strategies, and business impact.

Risk Report:
1. Cybersecurity Risk – Ransomware Exposure

Description:
Recent internal audits found outdated endpoint protection on ~18% of employee laptops. Two phishing simulation campaigns in January showed a 27% click‑through rate, significantly above the industry benchmark (12%).
Likelihood: High
Impact: Severe — Possible encryption of production systems, customer data breach, service outage 48–72 hours.
Key Findings:

Legacy antivirus engine no longer receiving signature updates.
VPN credentials stored locally without MFA enforcement.
SOC (Security Operations Center) identified unusual lateral movement attempts in early February.


Mitigation Plan:

Rollout of unified endpoint protection (UEM) by March 15.
Mandate MFA for all VPN access by March 1.
Conduct company-wide phishing training.


Residual Risk: Medium


2. Cloud Infrastructure Cost Overrun

Description:
Cloud expenditure exceeded budget by 19% in Q4 2025 due to unexpected data egress fees, idle VM clusters, and high‑volume analytics workloads.
Likelihood: Medium
Impact: Moderate — Could affect FY26 operating margin by 2–3%.
Key Findings:

34% of compute instances are underutilized (<20% CPU).
Data replication to multi‑region setup causing unnecessary egress charges.
No automated cost‑governance tooling in place.


Mitigation Plan:

Implement autoscaling for analytics workloads.
Remove unused VMs and switch from multi‑region to single‑region storage for non-critical workloads.
Adopt cloud cost visibility dashboards.


Residual Risk: Low–Medium


3. Third‑Party Vendor Stability Risk

Description:
Core payment gateway vendor reported financial instability and possible acquisition.
Likelihood: Medium
Impact: High — Service downtime may disrupt up to 38% of monthly transactions.
Key Findings:

Vendor delayed February SLAs twice due to staffing shortages.
API latency increased by 22% month-over-month.
Support tickets remain open for 11+ days on average.


Mitigation Plan:

Begin integration with alternate payment provider.
Create internal failover logic in transaction-processing microservices.
Add service credits and penalties into renewed contracts.


Residual Risk: Medium


4. Operational Risk – Data Quality Issues

Description:
Inconsistent customer data across CRM, billing, and support systems leading to billing errors and inaccurate forecasting.
Likelihood: High
Impact: Moderate — Potential revenue leakage of 3–5% monthly.
Key Findings:

12% mismatch between billing system and CRM customer records.
ETL pipelines lack automated validation checkpoints.
Customer service reports increased complaints due to incorrect invoices.


Mitigation Plan:

Launch Master Data Management (MDM) initiative.
Add validation rules and anomaly detection to ETL pipelines.
Monthly reconciliation of all customer data sources.


Residual Risk: Low
"""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt
)

display(Markdown("### Prompt Response:"))
display(Markdown(response.text))
```
