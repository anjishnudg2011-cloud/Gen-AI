```python
from google import genai
import os
from IPython.display import display, Markdown

# Configure API key (replace with your actual key or use environment variable)
GOOGLE_API_KEY = os.getenv("GEMINI_API_KEY", "YOUR_API_KEY_HERE")

# Initialize the client
client = genai.Client(api_key=GOOGLE_API_KEY)

# Set the model to use
MODEL_ID = "gemini-2.5-flash" 

# Define the prompt with system instructions and user query
prompt = """Objective: Analyse customer sentiment using the provided customer reviews and classify the last one accordingly.

Instructions: Think step‑by‑step and explain how you arrived at the answer.

Review: "The customer service was outstanding and the product works perfectly. Highly recommended!"
Sentiment: Positive

Review: "Terrible service. No one responded to my complaint and the product was defective."
Sentiment: Negative

Review: "I don't have strong feelings about this item. It’s neither great nor bad."
Sentiment: Neutral

Review: "I received the wrong product and it took forever to get support. Worst experience."
Sentiment: Negative

Review: "Packaging was decent. Product quality seems average so far."
Sentiment: Neutral

Review: "It meets the basic requirements but doesn’t stand out in any particular way."
Sentiment:
"""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt
)

display(Markdown("### Prompt Response:"))
display(Markdown(response.text))
```
