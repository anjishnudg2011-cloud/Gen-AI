```python
import requests
import os
from dotenv import load_dotenv
load_dotenv(override=True)  # override=True to reload variables if already loaded
# Get API key from environment variable
gemini_api_key = os.getenv("GEMINI_API_KEY")

if gemini_api_key is None:
    raise ValueError("GEMINI_API_KEY not found in .env file")

# Set up the API endpoint
url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-3-flash-preview:generateContent"

# Configure headers
headers = {
    "Content-Type": "application/json",
    "x-goog-api-key": gemini_api_key
}
```

```python
# Create the prompt payload
prompt = """Below is a Sample Loan Agreement document. Please think step-by-step and extract the key terms from it and present it in bullet format.
SAMPLE LOAN AGREEMENT
Date: [DD/MM/YYYY]
This Loan Agreement (“Agreement”) is made between:

Lender: [Lender Name], [Address]
Borrower: [Borrower Name], [Address]


1. Loan Amount
The Lender agrees to lend the Borrower ₹[Amount] (“Loan”).

2. Purpose
The Borrower shall use the Loan for [Purpose] only.

3. Interest
The Loan will carry interest at [X]% per annum, calculated on the outstanding principal.

4. Repayment Terms

Total payable = Principal + Interest
Repayment deadline: [DD/MM/YYYY]
Payment method: [UPI/Bank Transfer/Cash/Installments]
If installments: ₹[Amount] per [week/month]


5. Late Payment & Penalties

Late fee of [X]% per missed payment
Interest continues to accrue on overdue amounts
Lender may charge reasonable collection costs


6. Default
Default includes:

Failure to pay within [X] days of due date
Providing false information
Insolvency/bankruptcy
Breach of any clause of this Agreement

On default, the Lender may declare the entire Loan immediately due.

7. Security / Collateral (Optional)
This Loan is:

Unsecured, OR
Secured, with collateral described below:

[Item/Property]
Borrower confirms ownership and that it is free of other claims.



If Borrower defaults, Lender may take possession or enforce sale of the collateral.

8. Early Repayment
Borrower may prepay the Loan:

Without penalty, OR
Subject to a prepayment fee of [X]% of remaining principal

(Choose one)

9. No Transfer
Borrower cannot transfer this Loan obligation to another person without written consent from the Lender.

10. Communication
All notices or communication shall be made via:

Email: [Borrower email], [Lender email], OR
WhatsApp/SMS if mutually agreed


11. Privacy & Confidentiality
Both Parties agree not to disclose the terms of this Agreement to third parties, except as required by law.

12. Governing Law
This Agreement is governed by the laws of [State/Country].

13. Dispute Resolution
If a dispute arises:

Parties shall attempt amicable settlement within 7 days.
If unresolved, dispute will be settled by [Mediation/Arbitration/Court].
Jurisdiction: [City].


14. Amendments
Any change to this Agreement must be in written form and signed by both Parties.

15. Signatures
LENDER
Name: ________________________
Signature: _____________________
Date: _________________________
BORROWER
Name: ________________________
Signature: _____________________
Date: _________________________
WITNESS (optional)
Name: ________________________
Signature: _____________________"""
zero_shot_payload = {
    "contents": [
        {
            "parts": [
                {
                    "text": prompt
                }
            ]
        }
    ],
    "generationConfig": {
        "temperature": 0.8,
        "maxOutputTokens": 1500
    }
}
```

```python
# Send the request
response = requests.post(url, json=zero_shot_payload, headers=headers)

response_data = response.json()
print("\nUser prompt:\n",prompt)
print("\n" + "="*60 + "\n")
# Extract the text from the response
if 'candidates' in response_data:
    content = response_data['candidates'][0]['content']['parts'][0]['text']
    print("AI Response:\n")
    print(content.strip())
```
