```python
import requests
import os
from dotenv import load_dotenv
load_dotenv(override=True)  # override=True to reload variables if already loaded
# Get API key from environment variable
gemini_api_key = os.getenv("GEMINI_API_KEY")

if gemini_api_key is None:
    raise ValueError("GEMINI_API_KEY not found in .env file")

# Set up the API endpoint
url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-3-flash-preview:generateContent"

# Configure headers
headers = {
    "Content-Type": "application/json",
    "x-goog-api-key": gemini_api_key
}
# Create the prompt payload
prompt = """Objective: Integrate Google API with Python for real-time classification.

Persona: You are an INTENT CLASSIFIER for a customer-support chatbot.

Task: Classify the following chatbot messages into predefined intents. Use the following examples to help your classification.

Examples:
"user": "hi there"
"intent": "greeting"
"user": "where is my order 12345?"
"intent": "order_status"
"user": "i want a refund for my last purchase"
"intent": "refund_request"
"user": "please cancel my order asap"
"intent": "cancel_order"
"user": "do you have size 9 running shoes?"
"intent": "product_inquiry"
"user": "my card keeps getting declined"
"intent": "payment_issue"
"user": "human please"
"intent": "connect_agent"
"user": "thanks, bye"
"intent": "goodbye"
"user": "what are your store hours on sunday?"
"intent": "other"

"user": "hey, can you track order 778899?"
"intent":
"user": "i need my money back for a defective keyboard"
"intent":
"user": "connect me to a person please"
"intent":
"user": "payment failed twice—what should I do?
"intent":
"""

payload = {
    "contents": [
        {
            "parts": [
                {
                    "text": prompt
                }
            ]
        }
    ],
    "generationConfig": {
        "temperature": 0.3,
        "maxOutputTokens": 1500
    }
}
# Send the request
response = requests.post(url, json=payload, headers=headers)

response_data = response.json()
#print("\nUser prompt:\n",prompt)
#print("\n" + "="*60 + "\n")
# Extract the text from the response
if 'candidates' in response_data:
    content = response_data['candidates'][0]['content']['parts'][0]['text']
    print("AI Response:")
    print(content.strip())
```
