```python
from google import genai
import os
from IPython.display import display, Markdown

# Configure API key (replace with your actual key or use environment variable)
GOOGLE_API_KEY = os.getenv("GEMINI_API_KEY", "YOUR_API_KEY_HERE")

# Initialize the client
client = genai.Client(api_key=GOOGLE_API_KEY)

# Set the model to use
MODEL_ID = "gemini-2.5-flash" 

# Define the prompt with system instructions and user query
prompt = """Objective: Create audience-specific promotional content.

Task: Generate promotional content for a new credit card targeting three audiences: college students, frequent travellers, and senior citizens.

Persona: A marketing expert with a deep understanding of consumer behavior and preferences.

Instructions:
1. For college students, highlight the benefits of cashback on dining and entertainment, and a low annual fee.
2. For frequent travellers, emphasize the rewards points on travel expenses, airport lounge access, and travel insurance.
3. For senior citizens, focus on the ease of use, customer support, and benefits for healthcare expenses."""

response = client.models.generate_content(
    model=MODEL_ID,
    contents=prompt
)

display(Markdown("### Prompt Response:"))
display(Markdown(response.text))
```
