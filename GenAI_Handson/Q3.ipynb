```python
import requests
import os
from dotenv import load_dotenv
load_dotenv(override=True)  # override=True to reload variables if already loaded
# Get API key from environment variable
gemini_api_key = os.getenv("GEMINI_API_KEY")

if gemini_api_key is None:
    raise ValueError("GEMINI_API_KEY not found in .env file")

# Set up the API endpoint
url = "https://generativelanguage.googleapis.com/v1beta/models/gemini-3-flash-preview:generateContent"

# Configure headers
headers = {
    "Content-Type": "application/json",
    "x-goog-api-key": gemini_api_key
}
# Create the prompt payload
prompt = """Classify the intent of customer queries:

Query: "Hi, I want to open a new savings account with your bank. Could you please tell me what documents are required and whether I can complete the process online? Also, what is the minimum balance I need to maintain?"
Intent: Account Opening

Query: "I applied for a debit card last week but haven’t received any update yet. Can you check the status of my card dispatch? Also, please let me know how I can track the delivery."
Intent: Card Issue

Query: "I’m interested in applying for a personal loan. Could you share the eligibility criteria, interest rate, and required documents? Also, how long does the approval process usually take?"
Intent: Loan Inquiry

Query: "I am unable to log in to my net banking account. It keeps showing an authentication error. Can you help me reset my password or unlock my account?"
Intent: Net Banking

Query: "Hello, I want to know if I’m eligible for a personal loan. My monthly income is ₹55,000. Could you please tell me the interest rate, maximum loan amount I can get, and what documents I need to submit?"
Intent:"""
payload = {
    "contents": [
        {
            "parts": [
                {
                    "text": prompt
                }
            ]
        }
    ],
    "generationConfig": {
        "temperature": 0.2,
        "maxOutputTokens": 100
    }
}
# Send the request
response = requests.post(url, json=payload, headers=headers)

response_data = response.json()
#print("\nUser prompt:\n",prompt)
#print("\n" + "="*60 + "\n")
# Extract the text from the response
if 'candidates' in response_data:
    content = response_data['candidates'][0]['content']['parts'][0]['text']
    print("AI Response:")
    print(content.strip())
```
